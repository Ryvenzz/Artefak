```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('training_history30n.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['loss'], label=f'Training Loss Model 30 population', color='blue')
plt.plot(df['val_loss'], label=f'Validation Loss Model 30 population', linestyle='--', color='blue')

plt.title('Learning Curves for Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
# plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load history dari CSV
df_history = pd.read_csv('training_history30n.csv')

# Plot R
plt.figure(figsize=(10, 6))
plt.plot(df_history['r2_score'], label='Training R Model 30 Population', color='orange')
plt.plot(df_history['val_r2_score'], label='Validation R Model 30 Population', linestyle='--', color='orange')

plt.xlabel('Epoch')
plt.ylabel('R Score')
plt.title('Learning Curves for R')
# plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
