```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('training_history20n.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['loss'], label=f'Training Loss Model 20 population', color='green')
plt.plot(df['val_loss'], label=f'Validation Loss Model 20 population', linestyle='--', color='green')

plt.title('Learning Curves for Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
# plt.ylim(0, 2)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load history dari CSV
df_history = pd.read_csv('training_history20n.csv')

# Plot R²
plt.figure(figsize=(10, 6))
plt.plot(df_history['r2_score'], label='Training R² Model 20 population', color='orange')
plt.plot(df_history['val_r2_score'], label='Validation R² Model 20 population', linestyle='--', color='orange')

plt.xlabel('Epoch')
plt.ylabel('R² Score')
plt.title('Learning Curves for R²')
# plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
