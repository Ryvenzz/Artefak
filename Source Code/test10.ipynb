```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('training_history10.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['loss'], label=f'Training Loss Model 3', color='green')
plt.plot(df['val_loss'], label=f'Validation Loss Model 3', linestyle='--', color='green')

# plt.title('Learning Curves for All Models')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim(0, 1)
# plt.legend()
plt.grid(True)
plt.show()
```
