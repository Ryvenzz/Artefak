```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('konvergen_data10.csv')
df_2 = pd.read_csv('konvergen_data20.csv')
df_3 = pd.read_csv('konvergen_data30n.csv')
df_4 = pd.read_csv('konvergen_data40n.csv')
df_5 = pd.read_csv('konvergen_data50n.csv')
```

```python
iterasi = pd.Index([])  # Start with an empty Index to collect all unique column indices
iterasi = iterasi.union(df_1.columns.astype(int))
iterasi = iterasi.union(df_2.columns.astype(int))
iterasi = iterasi.union(df_3.columns.astype(int))
iterasi = iterasi.union(df_4.columns.astype(int))
iterasi = iterasi.union(df_5.columns.astype(int))

# Sorting the iteration values to be used on the x-axis
iterasi = sorted(iterasi)

# Plot values for all three models from the second row (index 1)
plt.plot(iterasi, df_1.iloc[1], label='pop 10', color='red')
plt.plot(iterasi, df_2.iloc[1], label='pop 20', color='green')
plt.plot(iterasi, df_3.iloc[1], label='pop 30', color='blue')
plt.plot(iterasi, df_4.iloc[1], label='pop 40', color='orange')
plt.plot(iterasi, df_5.iloc[1], label='pop 50', color='purple')

# Setting labels for the x and y axes
plt.xlabel('Iterasi')
plt.ylabel('Value')

# Adding a legend
plt.legend(loc='best')

# Display the plot
plt.show()
```
